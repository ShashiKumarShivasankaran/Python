#### 6. Find average age of those were pregnant and not pregnant

```python
avg_age_pregnant = clinic[clinic['Pregnant'] == 'yes']['Age'].mean()
avg_age_not_pregnant = clinic[clinic['Pregnant'] == 'no']['Age'].mean()

print(f'Avg_Age_Pregnant = {round(avg_age_pregnant)} \n Avg_Age_Not_Pregnant = {round(avg_age_not_pregnant)}')
```

```python
avg_age = clinic.pivot_table(values='Age', index='Pregnant', aggfunc='mean')
avg_age['Age'] = avg_age['Age'].round()
avg_age
```

#### 7. Find average blood pressure of those had drug reaction and did not had drug reaction

```python
avg_bp_drug = clinic[clinic['Drug_Reaction'] == 'yes']['BP'].mean()
avg_bp_not_drug = clinic[clinic['Drug_Reaction'] == 'no']['BP'].mean()

print(f'Avg_BP_Drug = {round(avg_bp_drug)} \n Avg_BP_Not_Drug = {round(avg_bp_not_drug)}')
```

```python
avg_bp = clinic.pivot_table(values='BP', index='Drug_Reaction', aggfunc='mean')
avg_bp['BP'] = avg_bp['BP'].round()
avg_bp
```

#### 8. Add new column name as ‘Age_group’ and those having age between 16 & 21 should be categorized as Low, more than 21 and less than 35 should be categorized as Middle and above 35 should be categorized as High.

```python
def Age_group(Age):
    if Age <= 21:
        return 'Low'
    elif Age >= 21 & Age <= 35:
        return 'Middle'
    else:
        return 'High'

clinic['Age_group'] = clinic['Age'].apply(Age_group)   
clinic
```
